```python
#importando librerias necesarias
import findspark
findspark.init()
from pyspark.sql import SparkSession
from pyspark import SparkContext
from pyspark import SparkConf
spark = SparkSession. \
builder. \
appName('UNIR'). \
config("spark.executor.memory", "1g"). \
config("spark.driver.memory", "2g"). \
config('spark.dirver.maxResultSize', '1g'). \
getOrCreate()

import pyspark	

from pyspark.sql import SparkSession, Row	

from pyspark.sql.types import MapType, StringType	

from pyspark.sql.functions import col	

from pyspark.sql.types import StructType,StructField, StringType	
```

```python
#importando base de datos
ruta_dataset = "hdfs://localhost:8020/data/covid/covid.csv"

df_covid = spark.read\
                .option("header", "true")\
                .option("inferSchema", "true")\
                .csv(ruta_dataset, inferSchema=True, header=True)

#imprimir el esquema para ver en que tipos de datos los esta reconociendo python y ver cuantos records son
df_covid.printSchema()
print('Total de registros:', df_covid.count())
```
